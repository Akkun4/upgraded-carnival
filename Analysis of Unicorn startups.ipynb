```python
## City wise count of companies

#sorted_df_2.groupby(['City']).agg(['count'])
#sorted_df_2.groupby(['City', 'Company']).agg(['count'])
sorted_df_3 = sorted_df_2['City'].value_counts(ascending=False)
df_value_counts = pd.DataFrame(sorted_df_3)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['City', 'Count'] # change column names
df_value_counts = df_value_counts.head(10)
```

```python
#Line graph - US Based Companies
#plt.plot(df_value_counts['City'], df_value_counts['Count'])  # Plot the chart
#plt.show()  
plt.bar(df_value_counts['City'], df_value_counts['Count'])
plt.title('City Vs Count')
plt.xlabel('City')
plt.ylabel('Count')

plt.xticks(df_value_counts['City'], df_value_counts['City'], rotation=45)
plt.show()
```

```python
#How many fintech comapny in ths US?

#sorted_df_2['Industry'] = sorted_df_2['Industry'].replace(['Artificial intelligence'],'Artificial Intelligence')
#sorted_df_2['Industry'] = sorted_df_2['Industry'].replace(['Finttech'],'Fintech')

sorted_df_2.loc[sorted_df_2["Industry"] == "Artificial intelligence", "Industry"] = 'Artificial Intelligence'
sorted_df_2.loc[sorted_df_2["Industry"] == "Finttech", "Industry"] = 'Fintech'


sorted_df_4 = sorted_df_2.groupby('Industry')[['Company']].count()
ind = sorted_df_4.index.to_list()

fig = plt.figure(figsize =(10, 10))

plt.pie(sorted_df_4['Company'], labels = ind, autopct='%1.1f%%')

plt.show()
```
